```python
inertia_df = pd.DataFrame({'Inertia': inertia}, index=[i for i in range(10,25)])
inertia_df
```

```python
inertia_df.plot(kind='line')
plt.ylabel('Sum of Squared Error(Inertia)')
plt.xlabel('Number of Cluster')
plt.title('Elbow Method')
```
