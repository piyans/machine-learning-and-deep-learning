## Import Library

```python
import pandas as pd
import numpy as np
```

## Input Dataset

```python
#Membaca data dan memasukkannya ke dalam bentuk Pandas Dataframe
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
```

```python
#Jalankan code untuk mengecek nama kolom yang tersedia
df.columns
df.head (5)
```

```python
#Hilangkan kolom yang dirasa tidak relevan dengan model (contoh: ID). None dapat diisi dengan nama-nama kolom yang akan digunakan.
#Contoh df = df[['X1','X2', 'Y']].copy()

#START CODE
df = df[['Geography', 'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'Exited']].copy()
df
```

## Preprocessing

```python
#Lakukan One-Hot Encoder untuk data categorical, dengan fungsi pandas get_dummies

#START CODE
df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
dummies = pd.get_dummies(df['Geography'], ['Gender'],drop_first = True)
df = pd.concat([df,dummies],axis = 1).drop('Geography', axis = 1).drop('Gender', axis = 1)
df
```

```python
#Pisahkan mana X (feature) dengan Y,
#Y adalah kolom "Exited"

#START CODE
#df = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/SC_HW1_bank_data.csv')
X = df.drop ('Exited', axis = 1)
y = df ['Exited']
```

```python
#Lakukan Scaler dan/atau Noermalisasi jika diperlukan
from sklearn.preprocessing import MinMaxScaler

#START CODE
scaler = MinMaxScaler()
X_transform = scaler.fit_transform(X)
```

```python
#Ini digunakan jika dilakukan scaler/Normalisas. Jika tidak, code ini bisa dilewat dan diganti dengan code yang ada di dalam komen
X_transform = pd.DataFrame(X_transform, columns = X.columns)
#X_transform = X.copy()
```

## Train-Test Split

```python
#Split menjadi train dan test dengan test_size 25%
#Tidak perlu mengubah code ini

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transform,y,test_size = 0.25,random_state = 123)
```

# Modeling

## Model1

### Soal :
Jelaskan secara Singkat Model pertama yang digunakan!

### Jawab:
Model pertama yang digunakan adalah logistic regresion, dengan parameter tol dan c. tol merupakan singkatan dari tolerance, merupakan parameter untuk mengukur convergence criterion. C merupakan parameter regulasi yang digunakan untuk mencegah overfitting dan mengontrol kompleksitas model. Logistic regresion menggunakan fungsi logistik (sigmoid) untuk mengubah keluaran linear menjadi probabilitas kelas positif (1) dan kelas negatif (0).

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

from sklearn.linear_model import LogisticRegression
#START CODE
model1 = LogisticRegression()
params = {
    'tol': [0.0001,0.001,0.01,0,1],
    'C':[0.001,0.01,0.1,1,10,100]
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model1,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train,y_train)
best_params = grid.best_params_
print("Hyperparameter Terbaik:", best_params)
```

```python
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("Array")
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score")
print(accuracy_score(y_test, y_pred))
```

## Model2
### Soal :
Jelaskan secara Singkat Model ke-2 yang digunakan!
### Jawab:
Model kedua yang digunakan adalah Support Vector Machine, dengan parameter C, gamma, dan kernal. Parameter C untuk mengontrol margin maksimum dan jumlah kesalahan klasifikasi pada data pelatihan. Parameter kernel digunakan untuk menentukan jenis fungsi kernel yang akan digunakan. SVM merupakan algoritma yang digunakan untuk tugas klasifikasi dan regresi, yang mampu menangani data yang kompleks dan pemisahan kelas yang tidak linear.

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

from sklearn.svm import SVC

#START CODE
model2 = SVC(kernel='linear',C=10)
params = {
    'C': [0.001, 0.01, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'rbf']
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model2,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train,y_train)
best_params = grid.best_params_
print("Hyperparameter terbaik Support Vector Machine:")
print(best_params)
```

```python
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("Array")
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score")
print(accuracy_score(y_test, y_pred))
```

## Model3
### Soal :
Jelaskan secara Singkat Model ke-3 yang digunakan!

### Jawab:
Pada model ketiga digunakan Random forest, dengan parameter n estimator, max depth, min sample split, dan min sample leaf. n estimator untuk mengontrol jumlah pohon keputusan yang akan dibangun. Max depth untukmembatasi tingkat kompleksitas setiap pohon, yang mana menunjukan kedalaman maksimum untuk setiap pohon keputusan. min sample split merupakan jumlah minimum sample yang diperlukan untuk membagi simpul dalam pohon keputusan. min sample leaf merupakan jumlah minimum yang diperlukan untuk menjadi daun (leaf) dalam pohon keputusan. Random forest merupakan algoritma ensemble dalam machine learning untuk klasifikasi dan regresi. Random forest menggabungkan prediksi dari beberapa model pohon keputusan (decision tree) untuk menghasilkan modek yang lebih stabil.

```python
#Pilih salah satu metode Machine Leaarning
#Model Machine Learning dapat dipanggil terlebih dahulu melalui library yang digunakan. Gunakan library scikit learn seperti pada contoh

#Contoh pemanggilan library dan penggunaannya dalam model

#from sklearn.linear_model import LogisticRegression
#model1 = LogisticRegression()
#params = { "tol": [0.1,0.01,0.001], 'C':[0.5,1.0,1.5,2.0]}

from sklearn.ensemble import RandomForestClassifier
#START CODE
model3 = RandomForestClassifier()
params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
#END CODE

#Lakukan parameter tuning sesuai hyperparameter yang dibutuhkan
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(
             estimator= model3,
             param_grid= params,
             scoring = 'accuracy',
             n_jobs = 10, # core cpu yang digunakan
             cv = 10 # 3-fold cross validation (artinya kita melakukan iterasi model sebanyak 3 kali)
            )

grid.fit(X_train, y_train)
best_params = grid.best_params_
print("Hyperparameter terbaik Random Forest:")
print(best_params)

grid.fit(X_train,y_train)
grid.best_params_
```

```python
#lakukan evaluasi
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_pred = grid.predict(X_test)

print(classification_report(y_test,y_pred))
print("Array")
print(confusion_matrix(y_test,y_pred))
print("Accuracy")
print(accuracy_score(y_test, y_pred))
```
